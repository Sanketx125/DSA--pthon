1. write a recursion function to calculate sum of first Nnatural numbers

```python
def sum(n):
    if n== 1:
        return 1
    else:
        s = n + sum(n-1)
    return s

sum(15)
        
    
```

2. wirte a recursion function to calculate sum of first N odd natural numbers

```python
def sum_odd(n):
    if n == 1:
        return 1
    else:
        if n%2 != 0:
            s = n + sum_odd(n-1)
            return s
        else:
            return sum_odd(n-1)

sum_odd(10)
```

3. wirte a recursion function to calculate sum of first N even natural numbers

```python
def sum_even(n):
    if n == 2:
        return 2
    else:
        if n%2 == 0:
            return n+ sum_even(n-1)
        else:
            return sum_even(n-1)

sum_even(10)
```

4. wirte a recursion function to calculate factorial of a number

```python
def fact(n):
    if n > -1:
        if n == 0:
            return 1
        else:
            s = n * fact(n-1)
        return s

fact(10)
```

5. 2. wirte a recursion function to calculate sum of squares of first N natural numbers

```python
def square_sum(n):
    if n > 0:
        if n == 1:
            return 1
        else:
            s = (n**2) + square_sum(n-1)
            return s

square_sum(5)
```
