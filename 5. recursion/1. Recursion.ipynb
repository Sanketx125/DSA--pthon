
# Recursion:

   * Function called it self is called recursion.
   * in recursion problem is solved in terms of problem itself
   * each time recursion function call to itself for little simpler version of original problem


Assignment:



1. write a recursion function to print N natural Numbers

```python
def recursion(n):
    if n>0:
        recursion(n-1)
        print(n)


recursion(5)
```

2. whrite a recursion function to print first N natural numbers in reverse order

```python
def natural(n):
    if n >0:
        print(n)
        natural(n-1)

natural(5)
```

3. whrite a recursion function to print first N odd natural number

```python
def odd(n):
    if n >0:
        odd(n-1)
        print(2*n-1)
        

odd(10)
```

4. whrite a recursion function to print first N even natural Numbers

```python
def even(n):
    if n>0:
        even(n-1)
        print(2*n)

even(10)
```

```python
def rev_odd(n):
    if n > 0:
        print(2*n-1)
        rev_odd(n-1)

rev_odd(10)
```
