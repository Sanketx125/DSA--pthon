## Queue using Linked-List


Assignment:



```python
class Node:
    def __init__(self, data= None, next= None):
        self.data = data
        self.next = next
```

```python

class Queue:
    def __init__(self, front= None, rear= None):
        self.front = front
        self.rear = rear
        self.item_count = 0

    def is_empty(self):
        return (self.rear == None)

    def enqueue(self, data):
        n = Node(data)
        if self.is_empty():
            self.front = n
        else: 
            self.rear.next = n
        self.rear = n
        self.item_count +=1

    def dequeue(self):
        if not self.is_empty():
            raise IndexError("Queue Underflow")
            
        elif self.front == self.rear:
            self.front = None
            self.rear = None
        else:
            self.front = self.front.next
        self.item_count -=1

    def get_front(self):
        if self.is_empty():
            raise IndexError("Queue Underflow")
        else:
            return self.front.data

    def get_rear(self):
        if self.is_empty():
            return self.rear.data
        else:
            raise IndexError("Queue Underflow")

    def get_size(self):
        return self.item_count
                        
        
```

```python
l = Queue()
```

```python
l.get_size()
```
