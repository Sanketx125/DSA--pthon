## Deque using list

   * Deque stands for double ended queue
   * deque is one of the variation of queue data structure.
   * the difference is only we can insert and delete the item from both ends

   * 


   * 

1.Operations of Deque:
-

   * insert front
   * insert rear
   * delete front
   * delete rear
   * get front
   * get rear
   * is_empty
   * size



Assignment:



```python
class Deque:
    def __init__(self):
        self.list = []

    def is_empyt(self):
        return len(self.list) ==0

    def insert_front(self,data):
        self.list.insert(0, data)

    def insert_rear(self, data):
        self.list.append(data)

    def delete_front(self):
        if self.is_empyt():
            raise IndexError("Deque underflow")
        else:
            self.list.pop(0)

    def delete_rear(self):
        if self.is_empyt():
            raise IndexError("Deque underflow")
        else:
            self.list.pop(-1)

    def get_front(self):
        if self.is_empyt():
            raise IndexError("Deque underflow")
        else:      
            return self.list[0]

    def get_rear(self):
        if self.is_empyt():
            raise IndexError("Deque underflow")
        else:
            return self.list[-1]

    def size(self):
        return len(self.list)
```
