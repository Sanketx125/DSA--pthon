## Queue using List

 * Que is linear data structure.
 * the working priciple of Queue is First in First out (FIFO)
 * it mainly two ends:
      * front   - (deletion)
      * rare    - (insertion)
   

1.operations on Queue:
-

   * enqueue    - insetion
   * dequeue    - deletion
   * is_empty   - checking is empty
   * get_front  - dispaly front item
   * get_rare   - dispaly rare item
   * size       - size of queue


Assignment:



```python
class Queue:
    def __init__(self):
        self.list = []

    def is_empty(self):
        return len(self.list) == 0

    def enqueue(self, data):
        self.list.append(data)

    def dequeue(self):
        if not self.is_empty():
            return self.list.pop(0)
        else:
            IndexError("Queue Underflow")

    def get_front(self):
        if not self.is_empty():
            return self.list[0]
        else:
            IndexError("Queue Underflow")   

    def get_rare(self):
        if not self.is_empty():
            return self.list[-1]
        else:
            IndexError("Queue Underflow")

    def get_size(self):
        return len(self.list)
    
```
