## Doubly Linked-List:
-------------------


    here,
         start  - start pointer
         pre    - previous nodes address
         data   - value which hold by the node
         next   - address of next node
-------------

1.Elementory Operations:
-
* Insertion
* Deletion
* Traversing
* Searching
* Checking for empty list


```python
class Node:
    def __init__(self, prev = None, data= None, next= None):
        self.prev = prev
        self.data = data
        self.next = next
```

```python

class DLL:
    
    def __init__(self, head=None):
        self.head = head

    def is_empty(self):
        return self.head is None

    def insert_at_start(self, data):
        n = Node(None, data, self.head)

        if self.is_empty():
            self.head = n
        else:
            self.head.prev = n
            self.head = n

    def insert_at_last(self, data):
        temp = self.head

        if self.is_empty():
            self.head = Node(None, data, None)
        else:
            while temp.next is not None:
                temp = temp.next
            n = Node(temp, data, None)
            temp.next = n

    def search(self, key):
        temp = self.head
        index = 0
        while temp is not None:
            if temp.data == key:
                print("Index: ", index)
                return temp
            temp = temp.next
            index += 1
        return None

    def insert_after(self, after, data):
        temp = self.search(after)
        if temp is not None:
            n = Node(temp, data, temp.next)
            if temp.next is not None:
                temp.next.prev = n
            temp.next = n

    def print_data(self):
        temp = self.head
        while temp is not None:
            print(temp.data, end="  ")
            temp = temp.next

    def delete_first(self):
        if not self.is_empty():
            self.head = self.head.next
            if self.head is not None:
                self.head.prev = None

    def delete_last(self):
        if self.head is None:
            return
        elif self.head.next is None:
            self.head = None
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.prev.next = None

    def delete_item(self, data):
        if self.head is None:
            return
        else:
            temp = self.head
            while temp is not None:
                if temp.data == data:
                    if temp.next is not None:
                        temp.next.prev = temp.prev
                    if temp.prev is not None:
                        temp.prev.next = temp.next
                    else:
                        self.head = temp.next
                    break
                temp = temp.next

    def __iter__(self):
        return DLLiterator(self.head)

```

```python
class DLLiterator:
    def __init__(self, head):
        self.current = head

    def __iter__(self):
        return self

    def __next__(self):
        if not self.current:
            raise StopIteration
        data = self.current.data
        self.current = self.current.next
        
        return data
```

```python
l = DLL()
```

```python
l.insert_at_last(10)
```

```python
l.print_data()
```

```python
l.insert_at_last(50)
```

```python
l.print_data()
```

```python
l.insert_after(10, 20)
```

```python
l.print_data()
```
