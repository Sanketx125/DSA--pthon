## Stack using Linked-List


Assignment:



```python
class Node:
    def __init__(self, data= None, next= None):
        self.data= data
        self.next= next
```

```python

class Stack:
    def __init__(self, top= None):
        self.item_count = 0
        self.top = top
        
    def is_empty(self):
        return self.top == None

    def push(self, data):
        n = Node(data, self.top)
        self.top = n
        self.item_count +=1


    def pop(self):
        if not self.is_empty():
            if self.top.next == None:
                self.top = None         
            else:
                data = self.top.data
                self.top = self.top.next
                self.item_count -=1
                return data
        raise IndexError("Stack is Empty")
        

    def peek(self):
        if not self.is_empty():
            return self.top.item
        raise IndexError("Stack is Empty")

    def get_size(self):
        return self.item_count

    
        
```
