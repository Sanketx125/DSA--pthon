## Stack:

* Stack is linear Data structure.
*  the working principle of Stack is FILO
*  FILO stands for First In Last Out

1.stack operations
-

*
    * Push     - insertion
    * Pop      - Deletion
    * is_empty - checking stack is empty or not
    * Peak     - checking Top element of stack
    * size     - checking size of stack

Assignment:



```python

class Stack:
    def __init__(self):
        self.list = []
        self.size = -1
        
    def is_empty(self):
        if not self.list:
            return len(self.list) == 0

    def push(self, data):
        self.list.append(data)
        self.size = self.size + 1

    def pop(self):
        if not self.is_empty():
            self.size = self.size - 1
            return self.list.pop

    def peek(self):
        return self.list[self.size]

    def get_size(self):
        return int(self.size +1)
        
```

```python
s = Stack()
```

```python
s.is_empty()
```

```python
s.push(10)
```

```python
s.is_empty()
```

```python
s.get_size()
```

```python
s.peek()
```
