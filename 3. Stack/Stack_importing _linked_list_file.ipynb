```python
class Stack:
    def __init__(self):
        self.list = Singly_linkedlist()
        self.size = 0

    def is_empty(self):
        return self.list.is_empty()

    def push(self, data):
        self.list.insert_at_start(data)
        self.size += 1

    def pop(self):
        if not self.is_empty():
            self.list.delete_head()
            self.size -= 1

    def get_size(self):
        return self.size

    def peek(self):
        if not self.is_empty():
            return self.list.head.item

    

    
    
```
