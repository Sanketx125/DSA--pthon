# Graph:

A graph is a collection of nodes (also called vertices) and edges that connect pairs of nodes. There are two main types of graphs:

* degree: (u):
   * degree is of node u written by deg(u). is the number of edges containing u.
   * if deg(u)= 0 then called it as isolated node




   * #### Directed Graph (Digraph):
        * Edges have a direction, going from one node to another.

   * #### Undirected Graph:
        * Edges do not have a direction, simply connecting two nodes.

## Types of Graphs

1. Weighted Graph: Edges have weights or costs associated with them.
2. Unweighted Graph: Edges do not have weights.
3. Cyclic Graph: Contains at least one cycle (a path that starts and ends at the same node).
4. Acyclic Graph: Does not contain any cycles.
5. Connected Graph: There is a path between every pair of nodes.
6. Disconnected Graph: Not all nodes are connected by paths.
7. Complete Graph: Every pair of distinct nodes is connected by a unique edge.
8. Bipartite Graph: Nodes can be divided into two disjoint sets such that no two nodes within the same set are adjacent.

## Graph Representations
1. Adjacency Matrix: A 2D array where the cell at row i and column j indicates the presence (and optionally, the weight) of an edge between nodes i and j.
2. Adjacency List: An array of lists where each list at index i contains the nodes that are adjacent to node i.
