```python
class Graph:
    def __init__(self, vno):
        self.vertex_count = vno
        self.adj_dict = {v: [] for v in range(vno)}

    def add_edge(self, u, v, weight = 1):
        if (0 <= u < self.vertex_count) and (0<= v <self.vertex_count):
            self.adj_dict[u].append((v, weight)) 
            self.adj_dict[v].append((u, weight)) 
        else:
            print("Invalid Vertex !")

    def remove_edge(self, u, v):
        if (0 <= u < self.vertex_count) and (0<= v <self.vertex_count):
            self.adj_dict[u] = [(vertex, weight) for vertex, weight in self.adj_dict[u] if vertex != v]
            self.adj_dict[v] = [(vertex, weight) for vertex, weight in self.adj_dict[v] if vertex != u]
        else:
            print("Invalid Vertex !")


    def has_edge(self, u, v):
        if (0 <= u < self.vertex_count) and (0<= v <self.vertex_count):
            return any( vertex == v for vertex, x in self.adj_dict[u])
        else:
            print("Invalid Vertex !")
            return False

    def print_adj_list(self):
        for vertex, n in self.adj_dict.items():
            print("V",vertex,": ",n)
            
        
        
```

```python
g = Graph(5)
```

```python
g.add_edge(0, 1)
g.add_edge(1,2)
g.add_edge(2,4)
g.add_edge(1,3)
g.add_edge(3,4)
```

```python
g.print_adj_list()
```
